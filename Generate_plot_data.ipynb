```python
import sys
import os
# dir = os.path.dirname(__file__)
# print('Get current working directory : ', dir + '/')
# sys.path.append(dir + '/')
import numpy as np
import random
import matplotlib.pyplot as plt
# import matplotlib as mpl
# plt.rcParams['text.usetex'] = True
from scipy.linalg import expm
from datetime import datetime 
```

```python
id_local = np.array([[1.,0],[0,1.]]) + 0*1j
sigma_x = np.array([[0,1.],[1.,0]]) + 0*1j
sigma_y = 1j*np.array([[0,-1.],[1.,0]]) + 0*1j
sigma_z = np.array([[1.,0],[0,-1.]]) + 0*1j
hadamard = 1.0/np.sqrt(2)*(sigma_x + sigma_z)+1j*0    
s = np.array([[1,0],[0,1j]])+1j*0
s_dagger = np.array([[1,0],[0,-1j]])+1j*0

M_u = np.array([[1,0],[0,0]]) + 0*1j # projector on spin-up, or |1>
M_d = np.array([[0,0],[0,1]]) + 0*1j # projector on spin-down, or |-1>
```

```python
def random_samples(size, rotations, bitstrings):
    x = random.sample(range(len(rotations)), size)
    rots = []
    bits = []
    for i in range(size):
        rots.append(rotations[x[i]])
        bits.append(bitstrings[x[i]])
        # print(rots, bits)
    return rots, bits
```

```python
def shadow_reconstruction(r, b, L):
    sums = 0.0
    rho = np.array([1])
    
    for i in range(len(r)):
        
        # m = np.array([1])
        
        if r[i] == 1:
            u = hadamard
        elif r[i] == 2:
            u = hadamard @ s_dagger
        else:
            u = id_local
            
        if b[i] == 1:
            p = M_u
        else:
            p = M_d
        
        m = 3 * u.conj().T @ p @ u - id_local
        # sums += np.trace(m @ sigma_z @ sigma_z)
        rho = np.kron(rho, m)
    return rho
```

```python
L_vals = np.arange(2, 13)
N_shadows = 1000 # Total number of shadows data stored
theta = 0.1
batchsize = 25 # Snapshots
N = 20 # Number of batches
```

```python
# Without phase-kickback

data = []
timevals = []
time_class = []


for i in range(len(L_vals)):
    
    L = L_vals[i]
    
    with open("L=" + str(L) +"_N="+str(N_shadows), "rb") as f:
        rotations = np.load(f)
        bitstrings = np.load(f)
        time = np.load(f)
    timevals.append(time)
        # G = np.load(f)
    print("L = ", L)

    G = np.zeros([2**L, 2**L], dtype = complex)
    G[0, -1] = -1j*L*np.exp(-1j * L * 0.1)
    G[-1, 0] = 1j*L*np.exp(1j * L * 0.1)
    # G = G * L * 1j
    # count = 0
    
    means = []
    start_time = datetime.now()
    for i in range(N):
        shots = []
        rots, bits = random_samples(batchsize, np.copy(rotations), np.copy(bitstrings))
        for j in range(batchsize):
            r = rots[j]
            b = bits[j]
            # count += 1
            # print(b)
            shots.append(np.trace(shadow_reconstruction(r, b, L) @ G @ G))
        means.append(np.mean(shots))
        print("K = "+str(i))
    end_time = datetime.now()
    data.append(means)
    time_class.append((end_time - start_time).total_seconds())
    
mean_vals = []
std_vals = []
for i in range(len(L_vals)):
    mean_vals.append(np.median(data[i]))
    std_vals.append(np.std(data[i]))
```

```python
# With phase-kickback

data_k = []
timevals_k = []
time_class_k = []
for i in range(len(L_vals)):
    
    L = L_vals[i]
    
    with open("L=" + str(L) +"_N="+str(N_shadows)+"_pkb", "rb") as f:
        rotations = np.load(f)
        bitstrings = np.load(f)
        time = np.load(f)
        # G = np.load(f)
    print("L = ", L)
    timevals_k.append(time)
    G = 1j * L * np.array([[0, -np.exp(-1j * L * theta)], [np.exp(1j * L * theta), 0.0]], dtype = complex)
    

    means = []
    start_time = datetime.now()
    for i in range(N):
        shots = []
        rots, bits = random_samples(batchsize, np.copy(rotations), np.copy(bitstrings))
        for j in range(batchsize):
            r = rots[j]
            b = bits[j]
       
            shots.append(np.trace(shadow_reconstruction(r, b, L) @ G @ G))
        means.append(np.mean(shots))
    end_time = datetime.now()
    data_k.append(means)
    time_class_k.append((end_time - start_time).total_seconds())
    
mean_vals_k = []
std_vals_k = []
for i in range(len(L_vals)):
    mean_vals_k.append(np.median(data_k[i]))
    std_vals_k.append(np.std(data_k[i]))
```

```python
with open("plot_data_snapshots=" + str(batchsize) +"_K="+str(N), "wb") as f:
    np.save(f, mean_vals)
    np.save(f, std_vals)
    np.save(f, mean_vals_k)
    np.save(f, std_vals_k)
    np.save(f, L_vals)
    np.save(f, timevals)
    np.save(f, timevals_k)
    np.save(f, time_class)
    np.save(f, time_class_k)
```
